```python
#A
#here we are considering data of dropped of ProsperRating (alpha) which is name by new_df in above problem.

condition = (new_df['LoanStatus'] == 'Completed') | (new_df['LoanStatus'] == 'Defaulted') |\
            (new_df['LoanStatus'] == 'Chargedoff')

df2 = new_df[condition]


def loanstatus(df2):
    if df2['LoanStatus'] == 'Chargedoff': 
        return 'Defaulted'
    else:
        return df2['LoanStatus']
df2['LoanStatus'] = df2.apply(loanstatus, axis =1)

df2['LoanStatus'].value_counts()
```

```python
#B
#list of dictionary initialization

list = {'categories':{1: 'Debt Consolidation', 2: 'Home Improvement', 3: 'Business', 6: 'Auto', 7: 'Other'}} 

#printing list
print(list)
```

```python
#here df2 is taken from above example
def category(j):
    d = {1:'Debt Consolidation', 2:'Home Improvement', 3:'Business', 6:'Auto', 7:'Other'}
    if j in d.keys():
        return d[j]
    else:
        return 'Other'
df2['ListingCategory (numeric)'] = df2['ListingCategory (numeric)'].apply(category)
df2['ListingCategory (numeric)'].value_counts()
```
